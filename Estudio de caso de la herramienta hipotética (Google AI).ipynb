```python
from docx import Document
from docx.shared import Pt, Inches
from docx.enum.text import WD_ALIGN_PARAGRAPH

# Crear documento
doc = Document()

# Configuración estilo APA 7
style = doc.styles["Normal"]
font = style.font
font.name = "Times New Roman"
font.size = Pt(12)

style.paragraph_format.line_spacing = 2
style.paragraph_format.space_after = 0

# =========================
# PORTADA
# =========================

doc.add_paragraph(
    "Visualización del Comportamiento del Modelo mediante The What-If Tool (Google AI): "
    "Un Análisis de Usabilidad e Interpretabilidad"
).alignment = WD_ALIGN_PARAGRAPH.CENTER

doc.add_paragraph("Jaime Lopez Fernandez").alignment = WD_ALIGN_PARAGRAPH.CENTER
doc.add_paragraph("Universidad / Máster en Inteligencia Artificial").alignment = WD_ALIGN_PARAGRAPH.CENTER
doc.add_paragraph("11 de febrero de 2026").alignment = WD_ALIGN_PARAGRAPH.CENTER

doc.add_page_break()

# =========================
# INTRODUCCIÓN
# =========================

doc.add_heading("Introducción", level=1)

doc.add_paragraph(
    "La Inteligencia Artificial Explicable (XAI) busca mejorar la transparencia y la confianza "
    "en los sistemas automatizados. The What-If Tool (WIT), desarrollada por Google AI, "
    "permite explorar el comportamiento de modelos de aprendizaje automático mediante "
    "manipulación interactiva de características sin necesidad de modificar el código del modelo."
)

# =========================
# ANÁLISIS
# =========================

doc.add_heading("Análisis", level=1)

doc.add_paragraph(
    "Durante la exploración de un conjunto de datos de riesgo crediticio, se modificaron "
    "variables como ingreso mensual, deuda total e historial de pagos. "
    "La herramienta permitió observar cambios probabilísticos en tiempo real."
)

doc.add_paragraph(
    "Se evidenció que el historial de pagos tenía mayor impacto que el ingreso absoluto, "
    "lo que demuestra la jerarquía implícita aprendida por el modelo."
)

# =========================
# CUADRO COMPARATIVO
# =========================

doc.add_heading("Cuadro Comparativo: WIT vs. SHAP vs. LIME", level=1)

doc.add_paragraph(
    "Comparación entre tres herramientas clave de Inteligencia Artificial Explicable."
)

table = doc.add_table(rows=6, cols=4)

headers = ["Criterio", "WIT", "SHAP", "LIME"]

for i in range(4):
    table.rows[0].cells[i].text = headers[i]

data = [
    ["Tipo de herramienta",
     "Interfaz visual interactiva",
     "Valores de Shapley (teoría de juegos)",
     "Aproximación lineal local"],

    ["Nivel de explicación",
     "Global e individual",
     "Global y local cuantitativa",
     "Principalmente local"],

    ["Fundamento teórico",
     "Visual analytics",
     "Teoría de juegos cooperativos",
     "Modelos lineales interpretables"],

    ["Facilidad de uso",
     "Alta",
     "Media",
     "Media"],

    ["Limitaciones",
     "No revela completamente la estructura interna",
     "Puede ser computacionalmente costoso",
     "Puede generar inestabilidad"]
]

for row_index, row in enumerate(data):
    for col_index, value in enumerate(row):
        table.rows[row_index + 1].cells[col_index].text = value

# =========================
# CONCLUSIÓN
# =========================

doc.add_heading("Conclusión", level=1)

doc.add_paragraph(
    "La visualización interactiva fortalece la interpretabilidad en modelos de aprendizaje "
    "automático. WIT facilita la exploración dinámica, mientras que SHAP y LIME ofrecen "
    "explicaciones matemáticamente fundamentadas. La integración de estos enfoques "
    "promueve mayor transparencia y confianza en sistemas de IA."
)

# =========================
# REFERENCIAS
# =========================

doc.add_heading("Referencias", level=1)

ref1 = doc.add_paragraph(
    "Norman, D. A. (2013). The design of everyday things (Revised and expanded ed.). Basic Books."
)
ref1.paragraph_format.left_indent = Inches(0.5)
ref1.paragraph_format.first_line_indent = Inches(-0.5)

ref2 = doc.add_paragraph(
    "Ribeiro, M. T., Singh, S., & Guestrin, C. (2016). Why should I trust you? "
    "Explaining the predictions of any classifier. Proceedings of the 22nd ACM SIGKDD "
    "International Conference on Knowledge Discovery and Data Mining, 1135–1144."
)
ref2.paragraph_format.left_indent = Inches(0.5)
ref2.paragraph_format.first_line_indent = Inches(-0.5)

# Guardar documento
doc.save("Informe_WhatIfTool_APA7.docx")

print("Documento generado correctamente.")
```
